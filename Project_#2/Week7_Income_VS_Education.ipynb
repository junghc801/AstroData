```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
income_arr = np.array([36700., 46400., 50200., 54200., 75100., 89200.])
education_arr = np.array([0, 1, 2, 3, 4, 5])
Education_levels = np.array([
"Less than high school",
"High school",
"College, No Degree",
"Associate",
"Bachelor",
"Master or Higher"
])
plt.plot(education_arr,income_arr, "o-")
plt.xlabel("Education Level")
plt.ylabel("Income [$]")
plt.xticks(education_arr, Education_levels, rotation=45)
plt.grid()
plt.show()
```

```python
income_error_arr = np.array([3110, 1070, 2000, 4130, 4110, 8910])


z_linear = np.polyfit(education_arr,income_arr,1)
print("Slope: " + str(z_linear[0]))
print("Intercept: " + str(z_linear[1]))
print()

z_quadratic = np.polyfit(education_arr,income_arr,2)
print("a: " + str(z_quadratic[0]))
print("b: " + str(z_quadratic[1]))
print("c: " + str(z_quadratic[2]))
print(str(round(z_quadratic[0], 2)) + "*X^2 + " + str(round(z_quadratic[1], 2)) + "*X + " + str(round(z_quadratic[2], 2)))
c_linear = z_linear[0] * education_arr + z_linear[1]
c_quadratic = z_quadratic[0] * education_arr**2 + z_quadratic[1] * education_arr + z_quadratic[2]



plt.errorbar(education_arr, income_arr, yerr=income_error_arr, marker="o", linestyle = "None", color = "black", label = "Data")
plt.plot(education_arr, c_linear, color = "red", label = "Linear fit")
plt.plot(education_arr, c_quadratic,   color = "blue", label = "Quadratic fit")
plt.xlabel("Education Level")
plt.ylabel("Income [$]")
plt.xticks(education_arr, Education_levels, rotation=45)
plt.grid()
plt.show()
```
